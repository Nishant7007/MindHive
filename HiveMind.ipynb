# **<h2>Task 1- ​ The world happiness index dataset​ for 2015, 2016 and 2017​</h2>**

**importing libraries**

```
import pandas as pd
import numpy as np
import warnings; warnings.simplefilter('ignore')
import os
import shutil
```

**Functions**

```
def addRegionColumn(df_2017):
  l = []
  for con in df_2017['Country']:
    try:
      reg = (df_2016[df_2016['Country'] == con]['Region']).tolist()[0]
    except:
      reg = 'NA'
    l.append(reg)
  df_2017['Region'] = l
  return df_2017
```

```
def formatCOlumns(df):
  df.columns = df.columns.str.replace(' ()' ,'.')
  return df
```

```
def getRelevantColumns(df):
  df = df[['Country','Region','Happiness.Score','Happiness.Rank']]
  return df
```

```
def getCountries(df):
  countries = (df[df['Happiness.Score']<5]['Country']).tolist()
  return countries
```

```
def task1_1(dfDict):
  HappinesScoreDict ={}
  for k,v in dfDict.items():
    year = k[3:]
    HappinesScoreDict[year] = getCountries(v)
    print('Countries with happiness score less than 5 for year ' + str(year)+ ' are:')
    print(HappinesScoreDict[year])
    print('')
```

```
def getUnhappiestCountry(df):
  df = df[df['Region']=='Sub-Saharan Africa']
  country = (df[df['Happiness.Rank'] == df['Happiness.Rank'].max() ]['Country']).tolist()[0]
  return country
```

```
def task1_2(dfDict):
  unhappiestCountryDict ={}
  for k,v in dfDict.items():
    year = k[3:]
    unhappiestCountryDict[year] = getUnhappiestCountry(v)
    print('Unhappiest country in SUb-Sahara Region for year ' + str(year)+ ' is:')
    print(unhappiestCountryDict[year])
    print('')
```

```
def getUnhappiestHappiestCountry(df,region):
  df = df[df['Region']==region]
  
  happiestCountry = df[df['Happiness.Rank'] == df['Happiness.Rank'].min()]
  happiestCountryName = happiestCountry['Country'].tolist()[0]
  happiestCountryRank = happiestCountry['Happiness.Rank'].tolist()[0]
  happiestCountryScore = happiestCountry['Happiness.Score'].tolist()[0]

  unHappiestCountry = df[df['Happiness.Rank'] == df['Happiness.Rank'].max()]
  unHappiestCountryName = unHappiestCountry['Country'].tolist()[0]
  unHappiestCountryRank = unHappiestCountry['Happiness.Rank'].tolist()[0]
  unHappiestCountryScore = unHappiestCountry['Happiness.Score'].tolist()[0]

  print('Happiest country is: '+str(happiestCountryName)+' with rank: '+str(happiestCountryRank))
  print('UnHappiest country is: '+str(unHappiestCountryName)+' with rank: '+str(unHappiestCountryRank))
  print(str(happiestCountryName) +' is '+str(happiestCountryScore/unHappiestCountryScore)+' times happier than '+str(unHappiestCountryName))
```

```
def task1_3(dfDict):
  regions = np.unique(dfDict['df_2015']['Region'])
  for k,v in dfDict.items():
      year = k[3:]
      print('FOR YEAR:',year)
      print('')
      for region in regions:
        print('FOR REGION:',region)
        getUnhappiestHappiestCountry(v,region)
        print('--------------------------------')
      print('===========================================')
```

```
def task1_4(dfDict):
  df = dfDict['df_2015']
  df = df[['Country','Happiness.Rank']]
  df.columns = ['Country','Happiness.Rank_2015']
  df_2017 = dfDict['df_2017']
  l = []
  for c in df['Country']:
    try:
      r = int((df_2017[df_2017['Country'] == c]['Happiness.Rank']).tolist()[0])
    except:
      r = np.nan
    l.append(r)
  df['Happiness.Rank_2017'] = l
  countries = df[df['Happiness.Rank_2015'] < df['Happiness.Rank_2017']]
  print('Countries that became unhappy between 2015 to 2017 are: ')
  print('')
  for ind in countries.index: 
     print(str(countries['Country'][ind]) +'from rank '+ str(df['Happiness.Rank_2015'][ind])+' in 2015 to rank '+ str(df['Happiness.Rank_2017'][ind])+' rank in 2017')
```

```
def task1_5(dfDict):
  df = dfDict['df_2015']
  df = df[['Country','Happiness.Score']]
  df.columns = ['Country','Happiness.Score_2015']
  df_2017 = dfDict['df_2017']
  l = []
  for c in df['Country']:
    try:
      r = int((df_2017[df_2017['Country'] == c]['Happiness.Score']).tolist()[0])
    except:
      r = np.nan
    l.append(r)
  df['Happiness.Score_2017'] = l
  val = (df['Happiness.Score_2015'] - df['Happiness.Score_2017']).max()
  x = df[(df['Happiness.Score_2015'] - df['Happiness.Score_2017']) == val]['Country'].tolist()[0]
  print('The country whose happiness decreased by the most amount is: ',x)
```

**Make a subdirectory for each label inside the train_data directory**

```
data = pd.read_csv('ucfTrainTestlist/classInd.txt', sep=" ", header=None)
folders = data[1].tolist()
root_path = 'train_data/'
for folder in folders:
  os.mkdir(os.path.join(root_path,folder))
```

**Iterating over all the trainlists (trainlist01.txt, trainlist02.txt, trainlist03.txt) and getting the labels and corresponding files and moving the files with those names in the corresponding labels subdirectories in the train_data directory**

```
folders = os.listdir('train_data')
for trainFiles in ['trainlist01.txt','trainlist02.txt','trainlist03.txt']:
  df = pd.read_csv('ucfTrainTestlist/'+str(trainFiles), sep='[/, ]', header=None)
  for folder in folders:
    folderPath = os.path.join('train_data',folder)
    files = (df[df[0] == folder][1]).tolist()
    for file in files:
      filePath = os.path.join('UCF-101',folder,file)
      try:
        shutil.move(filePath, folderPath)
      except:
        continue
```

**Iterating over each subdirectory in train_data and print number of files inside each subdirectory**

```
N = 0
for(dirpath, dirnames, filenames) in os.walk('train_data'):
    N_c = len(filenames)
    N += N_c
    print("Files in ", dirpath, N_c)
print("Total Files ",N)
```
